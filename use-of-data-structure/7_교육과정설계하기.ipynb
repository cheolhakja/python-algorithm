```python
from collections import deque

necessary = input()
num = int(input())
c = []
for i in range(num):
    tmp = input()
    c.append(tmp)

d = deque() #선언, 앞, 뒤 양쪽 방향에서 엘리먼트(element)를 추가하거나 제거할 수 있다.
for i in necessary: 
    d.append(i)
#-----input 건들일필요없음



def check(a: str):
    for i in a:
        if i in d:
            tmp = d.popleft()
            if tmp != i:
                print("#%d NO" %(i+1))
                exit(0)
    
    else:
        if(len(d) == 0):
            print("모든 필수과목을 이수했으니 성공")
        else:
            print("이수 안한 필수 과목이 있네 실패")
```
