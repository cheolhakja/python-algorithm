```python
from collections import deque

necessary = input()
num = int(input())
c = []
for i in range(num):
    tmp = input()
    c.append(tmp)

d = deque() #, ,    (element)    .
for i in necessary: 
    d.append(i)
#-----input 



def check(a: str):
    for i in a:
        if i in d:
            tmp = d.popleft()
            if tmp != i:
                print("#%d NO" %(i+1))
                exit(0)
    
    else:
        if(len(d) == 0):
            print("   ")
        else:
            print("     ")
```
